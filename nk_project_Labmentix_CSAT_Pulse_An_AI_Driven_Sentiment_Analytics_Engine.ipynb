### Dataset Loading

````python
# ==========================================
# 2. DATASET LOADING & INITIAL PREPARATION
# ==========================================

try:
    # 1. Load the dataset
    # Replace the path below if your file is in a different folder
    file_path = '/content/data/eCommerce_Customer_support_data.csv'
    df = pd.read_csv(file_path)

    # 2. Initial Data Inspection
    print(f"✅ Dataset Loaded: {df.shape[0]} rows and {df.shape[1]} columns.")

    # 3. Handling Nulls in Customer Remarks
    # We convert to string and fill NaNs to prevent errors during NLP tasks
    df['Customer Remarks'] = df['Customer Remarks'].astype(str).replace('nan', 'No Remarks')

    # 4. Feature Engineering: Sentiment Analysis
    # We do this NOW while the 'Customer Remarks' column is fully intact
    print("🔄 Calculating Sentiment Scores...")
    df['Sentiment_Score'] = df['Customer Remarks'].apply(lambda x: TextBlob(x).sentiment.polarity)

    # 5. Quick look at the updated data
    display(df[['Customer Remarks', 'Sentiment_Score']].head())
    print("✅ Sentiment Analysis complete. New column 'Sentiment_Score' created.")

except FileNotFoundError:
    print("❌ Error: The file was not found. Please check the file path.")
except Exception as e:
    print(f"❌ An unexpected error occurred: {e}")
````

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# ==========================================
# 3. DATA WRANGLING & FEATURE ENGINEERING
# ==========================================

# 1. Clean Column Names (Remove spaces and standardize)
df.columns = df.columns.str.strip().str.replace(' ', '_')
print("✅ Column names stripped and standardized.")

# 2. Identify and convert date columns dynamically
# Updated list to match standardized names (with underscores)
date_candidates = ['order_date_time', 'Issue_reported_at', 'issue_responded', 'Survey_response_Date']

for col in date_candidates:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
        print(f"✅ Converted to Datetime: {col}")

# 3. Calculate Handling_Min (Response Time)
# Using .abs() to ensure time is positive and handling missing dates
if 'issue_responded' in df.columns and 'Issue_reported_at' in df.columns:
    df['Handling_Min'] = (df['issue_responded'] - df['Issue_reported_at']).dt.total_seconds() / 60
    df['Handling_Min'] = df['Handling_Min'].fillna(df['Handling_Min'].median()).abs()
    print("✅ Calculated 'Handling_Min'. (Missing values filled with median)")

# 4. Text Cleanup (Prepare for Encoding later)
if 'Customer_Remarks' in df.columns:
    df['Customer_Remarks'] = df['Customer_Remarks'].fillna('No Remarks').astype(str)
    # Note: Sentiment_Score was already calculated in the previous step.
    print("✅ Customer Remarks cleaned and ready for visualization.")

display(df[['Issue_reported_at', 'issue_responded', 'Handling_Min']].head())
````

### What all manipulations have you done and insights you found?

The Manipulations (Cleaning the Mess)
Time Travel: I converted messy text timestamps into actual dates so the computer understands when things happened.

Filling Gaps: I filled in missing item prices and labeled empty customer remarks as "No Remarks" so the model doesn't trip over holes in the data.

Mind Reading (NLP): I used Sentiment Analysis to turn "I hate this" or "Great service" into a math score (-1 to +1) that the ANN can actually calculate.

Standardizing: I stripped hidden spaces from column names to stop those annoying KeyError messages.




The Insights (What the Data Says)
The "Silent" Gap: A large chunk of customers don't leave remarks; for those who do, their sentiment is the strongest predictor of the final CSAT score.

Speed vs. Satisfaction: There’s a "sweet spot" in handling time—too fast feels rushed, too slow feels lazy.

Shift Patterns: Support quality fluctuates based on the Agent Shift, suggesting burnout might be happening during specific hours.

### Based on your chart experiments, define three hypothetical statements from the dataset. In the next three questions, perform hypothesis testing to obtain final conclusion about the statements through your code and statistical testing.

Statement: There is a significant difference in the average Handling Time (minutes) between 5-star rated tickets and 1-star rated tickets.

Goal: To prove that speed is a statistically significant factor in achieving top-tier customer satisfaction

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

1. Research HypothesisNull Hypothesis ($H_0$): There is no significant difference in the mean handling time across different CSAT scores. (The timing doesn't affect the rating).Alternative Hypothesis ($H_1$): There is a significant difference in the mean handling time across different CSAT scores.

#### 2. Perform an appropriate statistical test.

````python
import scipy.stats as stats

# 1. Prepare the data: Group Handling_Min by CSAT_Score
# Dropping NaNs to ensure the test runs correctly
groups = [df[df['CSAT_Score'] == score]['Handling_Min'].dropna() for score in sorted(df['CSAT_Score'].unique())]

# 2. Perform One-Way ANOVA
f_stat, p_value = stats.f_oneway(*groups)

print("--- Hypothesis Test: ANOVA Results ---")
print(f"F-Statistic: {f_stat:.4f}")
print(f"P-Value: {p_value:.4e}")

# 3. Conclusion Logic
alpha = 0.05
if p_value < alpha:
    print("\nConclusion: Reject the Null Hypothesis.")
    print("There is a statistically significant difference in Handling Time across CSAT Scores.")
else:
    print("\nConclusion: Fail to Reject the Null Hypothesis.")
    print("There is no significant evidence that Handling Time differs by CSAT Score.")
````

##### Which statistical test have you done to obtain P-Value?

I performed the One-Way ANOVA (Analysis of Variance) test to calculate the P-value. This test determines if the differences in mean handling times across various CSAT scores are statistically significant.

##### Why did you choose the specific statistical test?

ANOVA was chosen because it is designed to compare the means of three or more groups (CSAT scores 1–5) simultaneously. It efficiently identifies if at least one group differs significantly without the need for multiple, less reliable t-tests.

````python
import re

# 1. DYNAMIC SEARCH: Automatically find your text column
text_targets = [c for c in df.columns if any(word in c.lower() for word in ['remark', 'text', 'cleaned'])]

if not text_targets:
    print("Error: No text column found!")
else:
    target_col = text_targets[0]

    def deep_clean(text):
        text = str(text)
        # Remove URLs (starting with http, https, or www)
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
        # Remove words that contain digits (e.g., 'order123', 'id99')
        text = re.sub(r'\w*\d\w*', '', text)
        # Remove any remaining standalone digits
        text = re.sub(r'\d+', '', text)
        # Remove extra whitespace left behind
        text = re.sub(r'\s+', ' ', text).strip()
        return text

    # Apply the deep clean
    df[target_col] = df[target_col].apply(deep_clean)

    print(f"Success! URLs and digits removed from '{target_col}'.")
    print(df[target_col].head())
````

````python
import nltk
from nltk.corpus import stopwords

# Ensure stopwords are downloaded
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# 1. DYNAMIC SEARCH: Automatically find your text column
text_targets = [c for c in df.columns if any(word in c.lower() for word in ['remark', 'text', 'cleaned'])]

if not text_targets:
    print("Error: No text column found!")
else:
    target_col = text_targets[0]

    def clean_stopwords_and_space(text):
        # Ensure text is string and lowercase
        text = str(text).lower()

        # 1. Remove Stopwords
        # Splits text into words, keeps only those NOT in stop_words
        words = [w for w in text.split() if w not in stop_words]

        # 2. Rejoin and Remove extra white spaces
        # " ".join() handles the spacing between words, .strip() handles the ends
        return " ".join(words).strip()

    # Apply the final cleaning
    df[target_col] = df[target_col].apply(clean_stopwords_and_space)

    print(f"Success! Stopwords and extra spaces removed from '{target_col}'.")
    print(df[target_col].head())
````
